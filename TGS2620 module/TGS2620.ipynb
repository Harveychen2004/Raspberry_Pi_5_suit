```python
from gpiozero import Button,MCP3008
from gpiozero.tools import absoluted, scaled
from signal import pause
import time
import pandas as pd
```

```python
TGS2620_AlcoholPin = MCP3008(channel=1)                # 定义烟雾传感器使用通道（AIN1)
TGS2620_DO = Button(13)                              # 烟雾传感器数字IO口(GPIO13)
```

```python
# 把x从[in_min, in_max]的范围映射到[out_min, out_max]的范围
def MAP(x, in_min, in_max, out_min, out_max):
    return (x - in_min) * (out_max - out_min) / (in_max - in_min) + out_min

def trans(x,a,b):
    return a*x**b
```

```python
def TGS2620_loop():
    data = []
    state = ["Dangerous", "Safe"]
    TGS2620_status = 1   # 定义状态值变量
    TGS2620_count = 0    # 定义计数器变量值
    try:    
        while True:
            current_time = time.strftime('%Y-%m-%d %H:%M:%S', time.localtime())
            
            TGS2620_InitVal = TGS2620_AlcoholPin.value # 读取AIN1上的模拟值
            TGS2620_TmpVal = round(MAP(TGS2620_InitVal,0,1,0,1000))                     
            # TGS2620_Alcohol = trans(TGS2620_Alcoholval, 0.03352, 0.552)
            print ('Alcohol Value: ', TGS2620_TmpVal)  
            # print ('Alcohol Value: ', TGS2620_Alcohol)  
            
            TGS2620_tmp = not TGS2620_DO.is_pressed  # 读取酒精传感器数字IO口值
            if TGS2620_tmp != TGS2620_status:     # 判断状态发生改变
                print(TGS2620_tmp)        # 打印函数，打印出酒精传感器信息
                TGS2620_status = TGS2620_tmp      # 把当前状态值设置为比较状态值，避免重复打印；
            if TGS2620_status == 0:                # 当检测到烟雾
                TGS2620_count += 1                 # 计数器值累计
            else:
                TGS2620_count = 0                  # 计数器赋0s
                
            data.append({'Time': current_time, "Initial Alcohol Value" : TGS2620_InitVal, "Processed Alcohol Value" : TGS2620_TmpVal, "State" : state[TGS2620_status], "Warning time period(s)" : TGS2620_count})
            time.sleep(1) # 延时1s
    except KeyboardInterrupt:
        df = pd.DataFrame(data)
        df.to_csv("data/TGS2620.csv", index=False)
```

```python
# 程序入口
if __name__ == '__main__':		
	try:
		TGS2620_loop() # 调用循环函数
	except KeyboardInterrupt:  	# 当按下Ctrl+C时，将执行destroy()子程序。
		print("Exit")  #资源释放
```
