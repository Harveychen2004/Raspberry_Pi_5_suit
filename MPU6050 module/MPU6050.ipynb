1. Import dependencies

```python
from mpu6050 import mpu6050
import time
import pandas as pd
```

2. I^2C address setting

```python
mpu6050_sensor = mpu6050(0x68)
```

3. Loop function

```python
def mpu6050_loop():
    data_save=[]
    try:
        while True:
            current_time = time.strftime('%Y-%m-%d %H:%M:%S', time.localtime())
            acceleration = mpu6050_sensor.get_accel_data()
            gyroscope = mpu6050_sensor.get_gyro_data()
            temp = mpu6050_sensor.get_temp()

            print("Accelerometer")
            print(acceleration)
            print("Gyroscope")
            print(gyroscope)
            print("Temp")
            print(temp)

            data_save.append({
                'Time': current_time,
                'Accelerometer_x': acceleration['x'],
                'Accelerometer_y': acceleration['y'],
                'Accelerometer_z': acceleration['z'],
                'Gyroscope_x': gyroscope['x'],
                'Gyroscope_y': gyroscope['y'],
                'Gyroscope_z': gyroscope['z'],
                'Temperature': temp
            })

            time.sleep(0.5)

    except KeyboardInterrupt:
        df=pd.DataFrame(data_save)
        df.to_excel("MPU6050_data.xlsx", index=False)
        print("Exit")
```

4. Main function

```python
if __name__=="__main__":
    try:
        mpu6050_loop()
    except KeyboardInterrupt:
        print("Exit")
```
